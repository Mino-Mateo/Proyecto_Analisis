```python
# Librerias
import csv
from pymongo import MongoClient

# Conexion MongoDB
mongo_client = MongoClient('localhost', 27017)
db_mongo = mongo_client['conciertos']

# Función para convertir CSV a JSON
def csv_to_json(csv_file_path):
    data = []
    with open(csv_file_path, "r", encoding="utf-8") as csv_archivo:
        csv_reader = csv.DictReader(csv_archivo)
        for row in csv_reader:
            data.append(row)
    return data

# Ruta de acceso del archivo CSV
archivo_csv = "../Bases_Datos/CSV/BF_Concert_2019-2020.csv"

# Convertir CSV a JSON
documentos_json = csv_to_json(archivo_csv)

# Guardado de archivos en MongoDB
for documento in documentos_json:
    # Guardar en MongoDB
    try:
        db_mongo['conciertos'].insert_one(documento)
    except Exception as e:
        print(f"Error al guardar en MongoDB")
```
