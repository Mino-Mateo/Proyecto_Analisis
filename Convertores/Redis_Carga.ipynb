```python
# Librerias
import csv
import redis
import json

# Conexion Redis
num_base = 1
redis_client = redis.StrictRedis(host='localhost', port=6379, db=num_base)

# Función para convertir CSV a JSON
def csv_to_json(csv_file_path):
    data = []
    with open(csv_file_path, "r", encoding="utf-8") as csv_archivo:
        csv_reader = csv.DictReader(csv_archivo)
        for row in csv_reader:
            data.append(row)
    return data

# Ruta de acceso del archivo CSV
archivo_csv = "../Bases_Datos/CSV/BF_Concert_2019-2020.csv"

# Convertir CSV a JSON
documentos_json = csv_to_json(archivo_csv)

# Guardado de archivos en Redis
contador = 1
for documento in documentos_json:
    # Asignar un identificador único a cada documento
    documento['_id'] = f"conciertos_{contador}"

    # Guardar en Redis
    try:
        redis_client.set(documento['_id'], json.dumps(documento))
        contador += 1
    except Exception as e:
        print(f"Error al guardar en Redis")
```
